```python
nfl_df['OFollow-Through'] = (nfl_df['PF'] / nfl_df_described.iloc[7,4]) * (nfl_df['W'] / nfl_df_described.iloc[7,0])
nfl_df['DFollow-Through'] = (nfl_df_described.iloc[3,5] / nfl_df['PA']) * (nfl_df['W'] / nfl_df_described.iloc[7,0])

# nfl_df.iloc[:,4:6].sum(axis=1) where axis = 1 specifies a horizontal summation 
#nfl_df.drop(columns=['Follow-Through'])
nfl_df_oft = nfl_df.sort_values('OFollow-Through', ascending=0)
nfl_df_dft = nfl_df.sort_values('DFollow-Through', ascending=0)
nfl_df['Follow-Through'] = ((nfl_df['OFollow-Through']) + (nfl_df['DFollow-Through']))/2
nfl_df_ft = nfl_df.sort_values('Follow-Through', ascending=0)
print(nfl_df_ft[['OFollow-Through','DFollow-Through','Follow-Through']])
#print(nfl_df_ft.loc['Dallas Cowboys','OFollow-Through'])
#print(nfl_df_ft)
```

```python
import random

def simulate_win(team1, team2):
    team1_oft = nfl_df_ft.loc[team1,'OFollow-Through'] + float(random.randrange(-25,25) / 100)
    team1_dft = nfl_df_ft.loc[team1,'DFollow-Through'] + float(random.randrange(-25,25) / 100)
    team2_oft = nfl_df_ft.loc[team2,'OFollow-Through'] + float(random.randrange(-25,25) / 100)
    team2_dft = nfl_df_ft.loc[team2,'DFollow-Through'] + float(random.randrange(-25,25) / 100)
    oft = (team1_oft - team2_dft)* 50
    dft = (team2_oft - team1_dft)* 50
    ft_rebuilt = 50 + oft + dft
    num = random.randrange(0, int(ft_rebuilt)-1)
    if num <= 25+(team1_oft + team2_dft):
        return True
    else:
        return False
        
def run_simulations(team1,team2,runs):
    team1_win = 0
    for run in range(runs):
        if simulate_win(team1, team2):
            team1_win += 1       
    #print(team1_win)
    return (team1_win / runs) * 100

def simulate_season(team):
    season = ['Tampa Bay Buccaneers', 'Cincinnati Bengals', 'New York Giants', 'Washington Football Team', 'Los Angeles Rams', 
    'Philadelphia Eagles', 'Detroit Lions', 'Chicago Bears', 'Green Bay Packers', 'Minnesota Vikings', 'New York Giants', 'Indianapolis Colts',
    'Houston Texans', 'Jacksonville Jaguars', 'Philadelphia Eagles', 'Tennessee Titans', 'Washington Football Team']
    record_w = 0
    record_l = 0
    for game in season:
        game_res = run_simulations(team, game, 1000)
        #print(game_res)
        if game_res > 50:
            record_w += 1
            print(f"The %s beat the %s!" % (team, game))
        else:
            record_l += 1
            print(f"The %s lost to the %s!" % (team, game))
    return record_w, record_l
        

print(f"The Cowboys went %s-%s in their simulations of the 2022 Season" % simulate_season('Dallas Cowboys'))
    
```
